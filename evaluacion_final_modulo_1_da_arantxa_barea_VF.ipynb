A lo largo de esta evaluación tendrás que crear una **clase** llamada **TiendaOnline** que cumpla los siguientes requisitos:

La clase TiendaOnline debe tener los siguientes **atributos**:

1. Inventario (lista de diccionarios): 
    Un atributo para almacenar los productos en el inventario.
    Cada producto debe ser representado como un diccionario con las siguientes claves:
    'nombre', 'precio', y 'cantidad'. 
Al principio deberá ser una lista vacía. Ejemplo de como debería ser: [{'nombre': 'Camisa', 'precio': 20, 'cantidad': 40}, {'nombre': 'Pantalón', 'precio': 30, 'cantidad': 30}]

2. ventas_totales (float): Un atributo para llevar un registro de las ventas totales de la tienda.
Inicializado con valor 0.


La clase TiendaOnline debe tener los siguientes **métodos**:

1. agregar_producto(self, nombre, precio, cantidad): 
Este método agrega un producto al inventario o actualiza su cantidad si ya existe. 
    - Debe recibir el nombre, precio y cantidad del producto como parámetros.
    - Itera a través del inventario y compara los nombres de los productos con el nombre proporcionado.
    -   Si el producto ya existe, actualiza la cantidad.
    -   Si no existe, agrega un nuevo producto al inventario.

2. ver_inventario(self): 
Muestra el inventario de productos con sus detalles.
    -   Utiliza un bucle for para recorrer el inventario.
    -   Imprime los detalles (nombre, precio, cantidad) de cada producto.
    -   Debería verse:
        -   Nombre: Camisa, Precio: $20, Cantidad: 50
        -   Nombre: Pantalón, Precio: $30, Cantidad: 30
        -   Nombre: Zapatos, Precio: $50, Cantidad: 40
        -   Nombre: Camisa, Precio: $20, Cantidad: 50

3. buscar_producto(self, nombre): 
Busca un producto en el inventario por nombre y muestra sus detalles si se encuentra. Debe recibir el nombre del producto como parámetro.
    - Utiliza un bucle for para recorrer el inventario.
    - Compara los nombres de los productos con el nombre proporcionado.
    -   Si se encuentra el producto, imprime sus detalles.
    -   Debe mostrar:
        -   Nombre: Camisa, Precio: $20, Cantidad: 40

4. actualizar_stock(self, nombre, cantidad): 
Actualiza el stock de un producto en el inventario. 
    - Debe recibir el nombre del producto y la cantidad a agregar o quitar como parámetros.
    - Utiliza un bucle for para recorrer el inventario.
    - Busca el producto por nombre.
    - Actualiza la cantidad según la entrada del usuario.
    -   Si el producto no esta en el inventario muestra un mensaje indicándolo.

5. eliminar_producto(self, nombre): 
Elimina un producto del inventario por nombre. 
    - Debe recibir el nombre del producto como parámetro.
    - Utiliza un bucle for para recorrer el inventario.
    - Busca el producto por nombre.
    - Elimina el producto del inventario si existe.
    -   Si el producto no esta en el inventario muestra un mensaje indicándolo.

6. calcular_valor_inventario(self): 
Calcula y muestra el valor total del inventario.
    - Utiliza un bucle for para calcular el valor total del inventario.
    - Itera a través del inventario y suma el valor de cada producto (precio x cantidad). Es decir, calcula el valor total del inventario. Ejemplo:
        - si tenemos 5 camisas que valen 5 euros y 10 calcetines que valen 1 euro, este método te tiene que devolver: 35 euros
            - valor_camisas = 5 * 5
            - valor_calcetines = 10 * 1
            - valor_camisas + valor_calcetines = 35

```python
#Definicion de la clase y sus métodos:
class TiendaOnline:
    def __init__(self): #Definimos los atributos de la clase en el método constructor
        self.inventario = [] #Lista de diccionarios vacía
        self.ventas_totales = 0.0 #Inicia en 0, tipo float(con decimal)

    def agregar_producto (self, nombre, precio, cantidad):
        nombre_formato = nombre.title()
        for producto in self.inventario: #Itero sobre una lista de diccionarios, producto es un diccionario.
            if producto['nombre'] == nombre_formato: #Accedo al valor del diccionario por la clave
                producto['cantidad'] += cantidad #Actualizo el valor actual de cantidad con la entrada de la función
                print(f"Inventario actualizado correctamente. Producto: {nombre_formato}, Cantidad: {producto['cantidad']} unidades.")
                return #una vez actualizado termina el bucle
        else: #Si no encuentra el producto lo agrega a la lista
            nuevo_producto = {'nombre': nombre_formato, 'precio': precio, 'cantidad': cantidad} #Defino nuevo producto, mismo formato que enunciado
            self.inventario.append(nuevo_producto) #Añado nuevo producto a la lista
            print(f"Producto añadido al inventario. Producto: {nombre_formato}, Precio: ${precio}, Cantidad: {cantidad} unidades")
            return
    
    def ver_inventario(self): #Como no recibe nada, primero genero las variables a imprimir
        for producto in self.inventario:
            nombre_formato = producto['nombre'].title()
            precio = producto['precio']
            cantidad = producto['cantidad']
            print(f'Nombre: {nombre_formato}, Precio: ${precio}, Cantidad: {cantidad}') #Formato definido en enunciado
    
    def buscar_producto(self, nombre): #Recibo de entrada el nombre de un producto e imprimo toda su información
        nombre_formato = nombre.title()
        for producto in self.inventario:
            if producto['nombre'] == nombre_formato:
                precio = producto['precio']
                cantidad = producto['cantidad']
                print(f"Nombre: {nombre_formato}, Precio: ${precio}, Cantidad: {cantidad}")
                return
        else:
            print(f'El producto {nombre_formato} no está registrado en el inventario')
            return
        
    def actualizar_stock(self, nombre, cantidad):
        nombre_formato = nombre.title()
        for producto in self.inventario:
            if producto['nombre'] == nombre_formato:
                producto['cantidad'] += cantidad
                print(f"Stock actualizado correctamente. Hay {producto['cantidad']} unidades de {nombre_formato}")
                return
        else:
            print(f'El producto {nombre_formato} no está registrado en el inventario')
            return

    def eliminar_producto(self, nombre):
        nombre_formato = nombre.title()
        for producto in self.inventario:
            if producto['nombre'] == nombre_formato:
                precio = producto['precio']
                cantidad = producto['cantidad']
                self.inventario.remove(producto)
                print(f"Producto eliminado correctamente. Has eliminado: {nombre_formato}, Precio: ${precio}, Cantidad: {cantidad} unidades.")
                return
        else:
            print(f'El producto {nombre_formato} no está registrado en el inventario')
            return      

    def calcular_valor_inventario(self):
        valor_inventario = 0  #Genero un acumulador a cero sobre que el ir sumando el valor de cada una de las prendas del inventario
        for producto in self.inventario:
            valor_producto = producto['precio'] * producto['cantidad'] #Primero calculo el valor de cada prenda
            valor_inventario += valor_producto #y luego lo incluyo y sumo uno a uno para tener un total que mostrar
        print(f"El valor total del inventario asciende a ${valor_inventario}")
        return
```

```python
#Bonus
#Para los ejercicios del bonus, creo una clase nueva que herede los métodos y atributos de la clase madre TiendaOnline.

class TiendaOnlineDos(TiendaOnline):
    
    def __init__(self): 
        super().__init__() #Ejecuta el constructor de la clase madre
        self.clientes = {} #nuevo atributo: diccionario vacío para llevar el registro de clientes

    def agregar_cliente(self, nombre_cliente, email):
        nombre_formato = nombre_cliente.title()
        if nombre_formato not in self.clientes:  #Si cliente no registrado, actualizo diccionario con update
            self.clientes.update({nombre_formato: {'email': email, 'compras': []}})
            print(f"Cliente registrado con éxito. Nombre: {nombre_formato}, Email: {email}")
        else:
            print(f"Cliente ya registrado en la base de datos.") #Si el cliente existe no sobreescribe, devuelve un mensaje de que el cliente ya existe en nuestro registro

    def ver_clientes(self):
        for nombre, informacion in self.clientes.items(): #Items() me devuelve tupla con par clave:valor. Esa tupla la desempaqueto en el bucle for
            email = informacion['email'] #el value de mi cliente es diccionario, accedo al email por su clave
            print(f"Nombre cliente: {nombre}, Email: {informacion['email']}")

    def realizar_compra(self):
        compra = True  #Defino la condición que debo cambiar en el bucle while para que no se ejecute infinitamente
        total_venta_carrito = 0 #Defino acumulador del total de ventas para sumar en caso de múltiples ventas
        carrito_final = {}
        self.ver_inventario() 
        while compra:
            try: #Valido en primer lugar las entradas, y en el except, manejo que la entrada de cantidad sea numérica
                producto_compra = input('Del inventario disponible, ¿qué producto desea comprar?').title()
                cantidad_compra = int(input(f"¿Cuántas unidades de {producto_compra}/s desea comprar?"))
            except ValueError:
                print(f"El valor introducido no es correcto. Introduce de nuevo la cantidad de compra.")
                continue
            else: #Cuando las entradas son válidas, valido que el producto existe y su stock
                producto_encontrado = False
                precio = 0
                cantidad_disponible = 0
                for producto in self.inventario: 
                    if producto_compra == producto['nombre']:
                        producto_encontrado = True
                        precio = producto['precio']
                        cantidad_disponible = producto['cantidad']
                        break
                if not producto_encontrado:
                        print(f"El producto {producto_compra} no se encuentra en el inventario, por favor, introduce de nuevo el producto que desea comprar.")
                        continue
                if cantidad_compra > cantidad_disponible:
                    print(f"Lo sentimos, no hay suficientes unidades de {producto_compra}, solo hay disponibles {cantidad_disponible} unidades. Comienza de nuevo con el proceso de compra.")
                    continue
                else: #cuando el producto existe y tiene unidades suficientes, permite realizar la compra
                    carrito_final[producto_compra] = {'precio': precio, 'cantidad': cantidad_compra} #registro el carrito de la compra
                    self.actualizar_stock(producto_compra, cantidad_compra * -1) #actualizo el inventario disponible
                    total_venta = precio * cantidad_compra #Calculo el valor del carrito
                    total_venta_carrito += total_venta #Voy añadiendo en caso de múltiples compras
                    nueva_compra = input('¿Desea añadir más productos al carrito?: Introduce Sí/No')
                    if nueva_compra.title() == 'Sí' or nueva_compra.title() == 'Si':
                        compra = True
                    else:
                        compra = False
        print(f"Compra realizada con éxito. El importe total de su compra asciende a ${total_venta_carrito}.")
        return carrito_final, total_venta_carrito #guardo el carrito y el importe para los próximos métodos

    def procesar_pago(self):
        while True:
            try:
                total_compra = float(input('Introduce el importe total de su compra: '))
                break
            except ValueError:
                print('El valor introducido no es correcto. Por favor, introduce de nuevo la información solicitada.')
                continue
        while True:                
            try:
                precio_pagado_cliente = float(input('Introduce el pago de su compra: '))
            except ValueError:
                print('El valor introducido no es correcto. Por favor, introduce de nuevo la información solicitada.')
                continue
            
            if precio_pagado_cliente == total_compra:
                print('Aquí tiene su compra. Esperamos verle de nuevo pronto.')
                break
            elif precio_pagado_cliente > total_compra:
                print(f"Aquí tiene su compra y su cambio de ${precio_pagado_cliente - total_compra}. Esperamos verle de nuevo pronto.")
                break
            else:
                print('No tiene dinero suficiente para procesar la compra. Por favor, introduce de nuevo el pago.')

    def registrar_compra(self, nombre_cliente, carrito):
        nombre_formato = nombre_cliente.title()
        if nombre_formato not in self.clientes:
            print('Para registrar la compra es necesario dar de alta al cliente en primer lugar.')
            email = input('Introduce el email del cliente para proceder con su alta.')
            self.agregar_cliente(nombre_formato, email)
            
        total_compra = 0
        for info_producto in carrito.values():
            total_compra += info_producto['precio'] * info_producto['cantidad']
        nueva_transaccion = {'productos': carrito, 'total': total_compra}
        self.clientes[nombre_formato]['compras'].append(nueva_transaccion)
        self.ventas_totales += total_compra
        print(f"Compra registrada para {nombre_formato}. Total: ${total_compra}")

    def ver_compras_cliente(self, nombre_cliente):
        nombre_formato = nombre_cliente.title()
        if nombre_formato in self.clientes:
            print(f"Nombre cliente: {nombre_formato}, Compras: {self.clientes[nombre_formato]['compras']}")
            return
        else: 
            print('Cliente no registrado en la base de datos')

    def calcular_ventas_totales(self):
        print(f"Las ventas totales de la tienda ascienden a: ${self.ventas_totales}") #Como ya acumule el total de todas las ventas en el metodo de registrar compra, imprimo el valor aquí
```
